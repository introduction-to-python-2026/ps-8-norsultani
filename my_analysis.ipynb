```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data directly from URL
url = 'https://gist.githubusercontent.com/vinaykusuma57/c663bd0a9f465740f3cdeff533b52f83/raw/global_coffee_data.csv'
df = pd.read_csv(url)
print("Loaded dataset!")
print(df.head())

# Save raw data
df.to_csv('raw_coffee_data.csv', index=False)
print("Saved raw data as raw_coffee_data.csv")

# Clean data
df_clean = df.dropna()
print(f"Rows before: {len(df)}, after cleaning: {len(df_clean)}")

# Visualization: Top 10 countries by consumption
plt.figure(figsize=(8,5))
sns.barplot(
    x='Country',
    y='Consumption (Metric Tons)',
    data=df_clean.sort_values('Consumption (Metric Tons)', ascending=False).head(10)
)
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Coffee Consumption")
plt.show()
```
